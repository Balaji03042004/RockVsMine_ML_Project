Importing the necessary library


```python
import  numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and data processing

```python
#loading the dataset to the pandas dataframe

sonar_data=pd.read_csv('/content/Copy of sonar data.csv',header=None)
```

```python

sonar_data.groupby(60).mean()
```

```python
#seperating data and label

X=sonar_data.drop(columns=60,axis=1)
Y=sonar_data[60]
```

```python
print(X)
print(Y)
```

```python
#  traning and test data
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
# we take the 10% of the data for the testing purpose by using the command of the test_size=0.1 or we can use the 0.2 for 20%
```

```python
print(x_train.shape,x_test.shape)
```

#model training by using the logistics regression

```python
model=LogisticRegression()
```

# training the logistics regression using the training data

```python
model.fit(x_train,y_train)
```

#model evaluation

```python
# check the accuracy on the training data

x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print(training_data_accuracy)
```

```python
# check the accuracy on the test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print(test_data_accuracy)
```
